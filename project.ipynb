```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = np.loadtxt('output/initial_collapsed_matrix.txt')
print('collapsed_matrix')
print('mean', np.mean(data)  )    # Mean
print('median', np.median(data))    # Median
print('std', np.std(data)   )    # Standard deviation
print('var', np.var(data)   )    # Variance
print('min', np.min(data)   )    # Minimum
print('max', np.max(data)   )    # Maximum
print('sum', np.sum(data)   )    # Sum
```

```python
data_raw = np.loadtxt('output/initial_raw_collapsed_matrix.txt')
print('raw collapsed_matrix')
print('mean', np.mean(data_raw)  )    # Mean
print('median', np.median(data_raw))    # Median
print('std', np.std(data_raw)   )    # Standard deviation
print('var', np.var(data_raw)   )    # Variance
print('min', np.min(data_raw)   )    # Minimum
print('max', np.max(data_raw)   )    # Maximum
print('sum', np.sum(data_raw)   )    # Sum
```

```python
data_hsr = np.loadtxt('output/initial_HSR_collapsed_matrix.txt')
print('HSR collapsed_matrix')
print('mean', np.mean(data_hsr)  )    # Mean
print('median', np.median(data_hsr))    # Median
print('std', np.std(data_hsr)   )    # Standard deviation
print('var', np.var(data_hsr)   )    # Variance
print('min', np.min(data_hsr)   )    # Minimum
print('max', np.max(data_hsr)   )    # Maximum
print('sum', np.sum(data_hsr)   )    # Sum
```

```python
data_hsr_raw = np.loadtxt('output/initial_HSR_raw_collapsed_matrix.txt')
print('HSR raw collapsed_matrix')
print('mean', np.mean(data_hsr_raw)  )    # Mean
print('median', np.median(data_hsr_raw))    # Median
print('std', np.std(data_hsr_raw)   )    # Standard deviation
print('var', np.var(data_hsr_raw)   )    # Variance
print('min', np.min(data_hsr_raw)   )    # Minimum
print('max', np.max(data_hsr_raw)   )    # Maximum
print('sum', np.sum(data_hsr_raw)   )    # Sum
```
